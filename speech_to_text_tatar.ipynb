```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("automatic-speech-recognition", model="AigizK/wav2vec2-large-mms-1b-tatar")
```

```python
# Load model directly
from transformers import AutoProcessor, AutoModelForCTC

processor = AutoProcessor.from_pretrained("AigizK/wav2vec2-large-mms-1b-tatar")
model = AutoModelForCTC.from_pretrained("AigizK/wav2vec2-large-mms-1b-tatar")
```

```python
# read your wav file into variable wav. For example:
from io import BytesIO
import numpy as np


wav = BytesIO()
with open('/content/1005.wav', 'rb') as f:
    wav.write(f.read())
wav.seek(0)

result = wav.read()
```

```python
asr = pipe(result, generate_kwargs={"language": "tatar", "max_new_tokens": 256}, return_timestamps='word')
```
